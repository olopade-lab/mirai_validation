```python
import pandas as pd
import numpy as np
import os
import cv2
```

```python
data_path= "../data/cleaned/mirai_full_cohort.csv"
metadata = pd.read_csv(data_path)
```

```python
metadata= metadata[:30]
```

```python
up_height = 2600
up_width = 2000
new_size = (up_width, up_height)
resized_dir = "/gpfs/data/huo-lab/Image/ojomoleye/projects/mirai_validation/batch_pngs/resized"
image_dir = "/gpfs/data/huo-lab/Image/ojomoleye/projects/mirai_validation/chimec_mammo_pngs"
for image_path in metadata['file_path']:
    # print('image path is', image_path)
    try: 
        image = cv2.imread(image_path)
        resized_up = cv2.resize(image, new_size, interpolation = cv2.INTER_LINEAR)
    #     cv2.imshow("resized up by defining width and height", resized_up)
        #save resized images
        resized_path = image_path.replace(image_dir, resized_dir, 1)
        os.makedirs(os.path.dirname(resized_path), exist_ok=True)
        print(resized_path)
        cv2.imwrite(resized_path, resized_up)
    
    except Exception as e:
        print("Problem opening", image_path, ":", e)
```
